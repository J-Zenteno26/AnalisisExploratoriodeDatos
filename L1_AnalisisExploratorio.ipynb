```python
# CREACIÓN DEL DATASET CON DATOS RANDOM, PERO DENTRO DE LO ESPERADO PARA UN ANÁLISIS CORRECTO.

import pandas as pd
import numpy as np

np.random.seed(42)
n = 300

edad = np.random.randint(18, 70, n)
genero = np.random.choice(["F", "M"], n)
visitas_web = np.random.poisson(10, n)

# Compras dependen un poco de visitas
compras = (visitas_web * 0.3 + np.random.normal(1, 1, n)).astype(int)
compras = np.clip(compras, 0, None)


monto_total = compras * np.random.normal(40000, 5000, n) # Monto depende directamente de compras
devoluciones = np.random.binomial(compras, 0.1)
calificacion = np.clip(5 - devoluciones + np.random.normal(0,1,n), 1, 5)
categoria_favorita = np.random.choice(["Tecnología", "Moda", "Hogar"], n)
suscripcion_newsletter = np.random.choice(["Sí", "No"], n)

data = pd.DataFrame({
    "edad": edad,
    "genero": genero,
    "visitas_web": visitas_web,
    "compras": compras,
    "monto_total": monto_total,
    "devoluciones": devoluciones,
    "calificacion": calificacion,
    "categoria_favorita": categoria_favorita,
    "suscripcion_newsletter": suscripcion_newsletter
})

# Guardamos data generado 
data.to_csv("comercioya.csv", index=False)
```
