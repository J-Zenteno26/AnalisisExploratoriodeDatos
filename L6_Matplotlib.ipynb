## Objetivo : Crear visualizaciones personalizadas y exportables. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Importamos la data
data = pd.read_csv("data/comercioya.csv")

fig, ax = plt.subplots(1, 2, figsize=(12,4))

# Scatter
ax[0].scatter(data["compras"], data["monto_total"], color="#4C72B0", alpha=0.6)
ax[0].set_title("Compras vs Monto Total")
ax[0].set_xlabel("Compras")
ax[0].set_ylabel("Monto Total")

# Histograma
ax[1].hist(data["monto_total"], bins=15, color="#55A868", edgecolor="black")
ax[1].set_title("Distribuci√≥n del Monto Total")
ax[1].set_xlabel("Monto Total")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Crear figura con 2x2 subplots
fig, ax = plt.subplots(2, 2, figsize=(14,8))

# Scatter + l√≠nea tendencia
ax[0,0].scatter(
    data["compras"],
    data["monto_total"],
    color="#4EC483",
    alpha=0.6,
    edgecolor="white"
)

# L√≠nea de tendencia
z = np.polyfit(data["compras"], data["monto_total"], 1)
p = np.poly1d(z)
x_sorted = np.sort(data["compras"])
ax[0,0].plot(x_sorted, p(x_sorted), color="#2F2F2F", linewidth=2)

ax[0,0].set_title("Compras vs Monto Total", fontweight="bold")
ax[0,0].set_xlabel("Compras")
ax[0,0].set_ylabel("Monto Total")
ax[0,0].grid(True, linestyle="--", alpha=0.4)

# Histograma
ax[0,1].hist(
    data["monto_total"],
    bins=15,
    color="#4EC483",
    edgecolor="black",
    alpha=0.7
)

ax[0,1].set_title("Distribuci√≥n del Monto Total", fontweight="bold")
ax[0,1].set_xlabel("Monto Total")
ax[0,1].set_ylabel("Frecuencia")

# Scatter Visitas vs Compras
ax[1,1].scatter(
    data["visitas_web"],
    data["compras"],
    color="#4EC483",
    alpha=0.6,
    edgecolor="white"
)

ax[1,1].set_title("Visitas Web vs Compras", fontweight="bold")
ax[1,1].set_xlabel("Visitas Web")
ax[1,1].set_ylabel("Compras")
ax[1,1].grid(True, linestyle="--", alpha=0.4)

# Boxplot
ax[1,0].boxplot(
    data["monto_total"],
    patch_artist=True,
    boxprops=dict(facecolor="#4EC483", alpha=0.6)
)

ax[1,0].set_title("Boxplot del Monto Total", fontweight="bold")
ax[1,0].set_ylabel("Monto Total")

# Ajuste autom√°tico del espacio entre subgr√°ficos
fig.tight_layout() # Evita superposici√≥n de t√≠tulos y etiquetas

# Ajustes generales
fig.suptitle("Resumen Visual del An√°lisis Exploratorio", fontsize=16, fontweight="bold")

plt.tight_layout()
plt.subplots_adjust(top=0.92)

# Guardar en alta calidad
plt.savefig("EDA_resumen_final.png", dpi=300, bbox_inches="tight")

plt.show()
```

Se construy√≥ una figura resumen que integra m√∫ltiples visualizaciones clave del an√°lisis exploratorio.<br> Se observa una fuerte relaci√≥n entre compras y monto_total, una distribuci√≥n asim√©trica del gasto, presencia de valores at√≠picos y una relaci√≥n positiva entre visitas_web y compras. <br>Esta visualizaci√≥n consolida los hallazgos principales del EDA.
